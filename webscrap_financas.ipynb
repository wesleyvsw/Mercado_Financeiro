```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
```

```python
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
url = "https://www.etf.com/etfanalytics/etf-finder"
driver.get(url)
```

## Achar todos os elementos necessários dentro do html do site - Pegando o número de páginas da tabela

```python
numero_paginas = driver.find_element('xpath','''/html/body/div[4]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/label[2]''')
numero_paginas = numero_paginas.text.replace("of ","")
numero_paginas =int(numero_paginas)
#print(numero_paginas.text) -----> assim eu coloco o texto que o selenium mandava
print(numero_paginas)
```

```python
#Lendo a tabela 1 do html
tabela = driver.find_element('xpath','''/html/body/div[4]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')

html_tabela = tabela.get_attribute('outerHTML')
tabela_final = pd.read_html(html_tabela)[0]
display(tabela_final)
```

```python
#Lendo todas as páginas
lista_de_tabelas_por_pagina = []

for x in range(0,numero_paginas):
    #/html/body/div[4]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]
    tabela = driver.find_element('xpath','''/html/body/div[4]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')
    html_tabela = tabela.get_attribute('outerHTML')
    tabela_final = pd.read_html(html_tabela)[0]
    lista_de_tabelas_por_pagina.append(tabela_final)
    botao_avancar_pagina = driver.find_element('xpath','''/html/body/div[4]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''') 
    driver.execute_script('arguments[0].click();',botao_avancar_pagina)
base_completa = pd.concat(lista_de_tabelas_por_pagina)
display(base_completa)
```
